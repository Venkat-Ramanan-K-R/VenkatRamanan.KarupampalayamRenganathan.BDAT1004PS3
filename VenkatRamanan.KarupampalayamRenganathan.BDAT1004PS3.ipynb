```python
#step1:

import pandas as pd

#step 2 and step 3:

users = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user", sep = "|")
users

```

```python
#step 4:

users.groupby("occupation")["age"].mean()

```

```python
#step 5:

#value_counts returns count of unique values. However, if the paramteter "normalize" set to True, it will return in
#relative freq.

users.groupby("occupation")["gender"].value_counts(normalize=True)

#multipying by 100 to get percentage

usr=users.groupby("occupation")["gender"].value_counts(normalize=True)*100

#now we have to get the ratio of male in all occupation category.

usr=pd.DataFrame(usr)

usr.rename(columns={"gender":"percent"}, inplace=True)

usr=pd.DataFrame(usr.reset_index(level=["occupation","gender"]))

usr[usr["gender"] == "M"].sort_values(by="percent",ascending=False)
```

```python
#Step 6. For each occupation, calculate the minimum and maximum ages

users.groupby("occupation").age.agg([max,min])
```

```python
# Step 7. For each combination of occupation and sex, calculate the mean age

users.groupby(["occupation","gender"]).age.mean()
```

```python
#Step 8. For each occupation present the percentage of women and men


users.groupby("occupation")["gender"].value_counts(normalize=True)

#multipying by 100 to get percentage

usr=users.groupby("occupation")["gender"].value_counts(normalize=True)*100

#now we have to get the ratio of male in all occupation category.

usr=pd.DataFrame(usr)
usr.rename(columns={"gender":"percent"}, inplace=True)
usr=pd.DataFrame(usr.reset_index(level=["occupation","gender"]))
usr
```

```python
#step1:

import pandas as pd
import datetime
```

```python
#step 2 & 3

data=pd.read_csv("wind.txt", sep = "\s+", parse_dates = [[0,1,2]])
data
```

```python
#Step 4. Year 2061? Do we really have data from this year? Create a function to fix it and apply it

def fix_century(x):
  yr = x.year - 100 if x.year > 1989 else x.year
  return datetime.date(yr, x.month, x.day)

data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(fix_century)

data
```

```python
# Step 5. Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns].

data["Yr_Mo_Dy"] = pd.to_datetime(data["Yr_Mo_Dy"])

data = data.set_index('Yr_Mo_Dy')

data
```

```python
# Step 6. Compute how many values are missing for each location over the entire record.
#They should be ignored in all calculations below.

data.isnull().sum()
```

```python
#Step 7. Compute how many non-missing values there are in total
data.notnull().sum()
```

```python
#Step 8. Calculate the mean windspeeds of the windspeeds over all the locations and all the times.

round(data.sum().sum() / data.notna().sum().sum(),3)
```

```python
#Step 9. Create a DataFrame called loc_stats and calculate the min, max and mean 
#windspeeds and standard deviations of the windspeeds at each location over all the days

loc_stats=data.describe(percentiles=[])
loc_stats
```

```python
# Step 10. Create a DataFrame called day_stats and calculate the min, max and mean
#windspeed and standard deviations of the windspeeds across all the locations at each day.

day_stats = pd.DataFrame()

# this time we determine axis equals to one so it gets each row.
day_stats['min'] = data.min(axis = 1) # min
day_stats['max'] = data.max(axis = 1) # max 
day_stats['mean'] = data.mean(axis = 1) # mean
day_stats['std'] = data.std(axis = 1) # standard deviations

day_stats
```

```python
#Step 11. Find the average windspeed in January for each location. Treat January 1961 and January 1962 both as January.

data.loc[data.index.month == 1].mean()
```

```python
#Step 12. Downsample the record to a yearly frequency for each location

data.groupby(data.index.to_period('Y')).mean()
```

```python
#Step 13. Downsample the record to a monthly frequency for each location

data.groupby(data.index.to_period('M')).mean()
```

```python
#Step 14. Downsample the record to a weekly frequency for each location.

data.groupby(data.index.to_period('W')).mean()
```

```python
#Step 15. Calculate the min, max and mean windspeeds and standard deviations of the
#windspeeds across all locations for each week (assume that the first week starts on
#January 2 1961) for the first 52 weeks.

weekly_dt = data.resample('W').agg(['min','max','mean','std'])
weekly_dt.loc[weekly_dt.index[1:53], "RPT":"MAL"] .head(10)
```

# Question 6
Create a line plot showing the number of marriages and divorces per capita in the
U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!


```python
import pandas as pd
import matplotlib.pyplot as plt

us_marriage_divorce_data = pd.read_csv('us-marriages-divorces-1867-2014.csv')

plt.plot(us_marriage_divorce_data.Year, us_marriage_divorce_data['Marriages'], label='Marriages')
plt.plot(us_marriage_divorce_data.Year, us_marriage_divorce_data['Divorces'], label='Divorces')

plt.title("US Marriage Divorce between 1867-2014")
plt.xlabel("Years")
plt.ylabel("No. of Marriage & Divorces per capita");

plt.legend()
```

# Question 7 Python

Create a vertical bar chart comparing the number of marriages and divorces per capita in the U.S. between 1900, 1950, and 2000.

```python
import pandas as pd
import matplotlib.pyplot as plt
fil= pd.read_csv('us-marriages-divorces-1867-2014.csv')

data = fil[(fil.Year == 2000) | (fil.Year == 1900) | (fil.Year == 1950)]
data.plot(kind= 'bar', x='Year', y=['Marriages_per_1000','Divorces_per_1000'], label=['Marriages','Divorces'])
plt.title("No. of marriages and divorces between in the years of 1900, 1950, and 2000.")
plt.ylabel('Marriages and Divorces per capita')
plt.legend()
```

# Question 9
Create a pie chart showing the fraction of all Roman Emperors that were
assassinated.
Make sure that the pie chart is an even circle, labels the categories, and shows the
percentage breakdown of the categories.


```python
import pandas as pd
import matplotlib.pyplot as plt

fil = pd.read_csv("roman-emperor-reigns.csv")

dead_emperors = fil.groupby('Cause_of_Death')['Emperor'].count()

dead_emperors.plot.pie(y='Cause_of_Death', figsize=(7, 7),autopct='%1.1f%%')

plt.title('Causes of death of Roman Emperors')
plt.ylabel('')
```
